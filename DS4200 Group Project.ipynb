```python
alt.data_transformers.disable_max_rows()
```

```python
correlation_columns_adjusted = [
    'Stress_Level', 'CGPA', 'Sleep_Quality', 'Physical_Activity',
    'Diet_Quality', 'Social_Support', 'Financial_Stress',
    'Extracurricular_Involvement', 'Semester_Credit_Load', 'Age'
]

data_adjusted = df.copy()
mapping_quality = {'Poor': 1, 'Average': 2, 'Good': 3}
data_adjusted['Sleep_Quality'] = data_adjusted['Sleep_Quality'].map(mapping_quality)
data_adjusted['Physical_Activity'] = data_adjusted['Physical_Activity'].map({'Low': 1, 'Moderate': 2, 'High': 3})
data_adjusted['Diet_Quality'] = data_adjusted['Diet_Quality'].map(mapping_quality)
data_adjusted['Social_Support'] = data_adjusted['Social_Support'].map({'Low': 1, 'Moderate': 2, 'High': 3})
data_adjusted['Extracurricular_Involvement'] = data_adjusted['Extracurricular_Involvement'].map({'Low': 1, 'Moderate': 2, 'High': 3})
```

```python
#Correlation with Stress Level (Female, Age 18-25)
correlation_columns_adjusted = [
    'Stress_Level', 'CGPA', 'Sleep_Quality', 'Physical_Activity',
    'Diet_Quality', 'Social_Support', 'Financial_Stress',
    'Extracurricular_Involvement', 'Semester_Credit_Load', 'Age'
]

data_adjusted_female_18_25 = data_adjusted[
    (data_adjusted['Gender'] == 'Female') & 
    (data_adjusted['Age'] >= 18) & 
    (data_adjusted['Age'] <= 25)
]

correlation_matrix_female_18_25 = data_adjusted_female_18_25[correlation_columns_adjusted].corr()

correlation_data_female_18_25 = correlation_matrix_female_18_25[['Stress_Level']].reset_index()

correlation_data_female_18_25.columns = ['Variable', 'Correlation_with_Stress_Level']

correlation_data_female_18_25 = correlation_data_female_18_25[
    correlation_data_female_18_25['Variable'] != 'Stress_Level'
]

chart = alt.Chart(correlation_data_female_18_25).mark_bar().encode(
    x=alt.X(
        'Correlation_with_Stress_Level:Q', 
        title='Correlation with Stress Level', 
        scale=alt.Scale(domain=(-0.1, 0.1))
    ),
    y=alt.Y('Variable:O', sort='-x'),
    color=alt.condition(
        alt.datum.Correlation_with_Stress_Level > 0,
        alt.value("blue"),
        alt.value("red")
    )
).properties(
    title="Correlation with Stress Level (Female, Age 18-25)"
)
chart.save("female_correlation_stress_level.html")
chart
```

```python
#Correlation with Stress Level (Male, Age 18-25)
correlation_columns_adjusted = [
    'Stress_Level', 'CGPA', 'Sleep_Quality', 'Physical_Activity',
    'Diet_Quality', 'Social_Support', 'Financial_Stress',
    'Extracurricular_Involvement', 'Semester_Credit_Load', 'Age'
]

data_adjusted_male_18_25 = data_adjusted[
    (data_adjusted['Gender'] == 'Male') & 
    (data_adjusted['Age'] >= 18) & 
    (data_adjusted['Age'] <= 25)
]

correlation_matrix_male_18_25 = data_adjusted_male_18_25[correlation_columns_adjusted].corr()

correlation_data_male_18_25 = correlation_matrix_male_18_25[['Stress_Level']].reset_index()

correlation_data_male_18_25.columns = ['Variable', 'Correlation_with_Stress_Level']

correlation_data_male_18_25 = correlation_data_male_18_25[
    correlation_data_male_18_25['Variable'] != 'Stress_Level'
]

chart = alt.Chart(correlation_data_male_18_25).mark_bar().encode(
    x=alt.X(
        'Correlation_with_Stress_Level:Q', 
        title='Correlation with Stress Level', 
        scale=alt.Scale(domain=(-0.1, 0.1)) 
    ),
    y=alt.Y('Variable:O', sort='-x'),
    color=alt.condition(
        alt.datum.Correlation_with_Stress_Level > 0,
        alt.value("blue"), 
        alt.value("red")    
    )
).properties(
    title="Correlation with Stress Level (Male, Age 18-25)"
)
chart.save("male_correlation_stress_level.html")
chart
```

```python
import plotly.graph_objects as go

courses = data_adjusted_female_18_25["Course"].unique()
stress_levels = sorted(data_adjusted_female_18_25["Stress_Level"].unique())

fig = go.Figure()

course_stress_level = data_adjusted_female_18_25.groupby("Course")["Stress_Level"].mean().reset_index()
fig.add_trace(go.Bar(
    x=course_stress_level["Course"],
    y=course_stress_level["Stress_Level"],
    name="Average Stress Level by Course",
    visible=True  
))

for course in courses:
    course_data = data_adjusted_female_18_25[data_adjusted_female_18_25["Course"] == course]
    for stress_level in stress_levels:
        level_data = course_data[course_data["Stress_Level"] == stress_level]
        fig.add_trace(go.Box(
            x=level_data["Stress_Level"].astype(str),  
            y=level_data["CGPA"],
            name=f"Stress Level {stress_level}", 
            visible=False,
            boxmean="sd"
        ))

dropdown_buttons = []

all_visibility = [True] + [False] * (len(courses) * len(stress_levels))
dropdown_buttons.append(dict(
    label="All Courses",
    method="update",
    args=[{"visible": all_visibility},
          {"title": {
                      "text": "Average Stress Level by Course(Female, Age 18-25)",
                      "x": 0.5,
                      "xanchor": "center"
                  },
           "xaxis": {"title": "Course", "type": "category"},
           "yaxis": {"title": "Stress Level"}}]
))

for i, course in enumerate(courses):
    visibility = [False] * (1 + len(courses) * len(stress_levels))
    for j in range(len(stress_levels)):
        visibility[1 + i * len(stress_levels) + j] = True 
    dropdown_buttons.append(dict(
        label=course,
        method="update",
        args=[{"visible": visibility},
              {"title": {
                      "text": f"CGPA by Stress Level for {course}",
                      "x": 0.5,
                      "xanchor": "center"
                  },
               "xaxis": {"title": "Stress Level", "type": "category"},
               "yaxis": {"title": "CGPA"}}]
    ))

fig.update_layout(
    updatemenus=[
        {
            "buttons": dropdown_buttons,
            "direction": "down",
            "showactive": True,
            "x": 0.1,
            "y": 1.2,
        },
    ],
    title= {
                      "text": "The Relationship between Stress Level and Course[Female, Age 18-25]",
                      "x": 0.5,
                      "xanchor": "center"
                  },
    xaxis_title="Course",
    yaxis_title="Stress Level" 
)

fig.write_html("interactive_boxplot_courses_fixed_axis.html")
fig.show()
```

## Visualization on Anxiety Score:
I created one interactive and one static plot for this relationship. 

### Heatmap (Static): 
This static plot is a heatmap of the following features:
 - Sleep Quality
 - Physical Activity
 - Anxiety Score
 
The table shows the relationship between sleep quality (poor, good, average) with physical activity (high, low, moderate). The hue gradient is determined by each relationship's average anxiety score. 

```python
## Create the heatmap table for sleep quality vs physical activity in conjuction with anxiety score.
pivot_table = df.pivot_table(values = 'Anxiety_Score', 
                               index = 'Sleep_Quality', 
                               columns = 'Physical_Activity', 
                               aggfunc = 'mean')

## Plot heatmap.
plt.figure()
sns.heatmap(pivot_table, annot = True, fmt = ".2f", cbar_kws = {'label': 'Average Anxiety Score'})
plt.title("Average Anxiety Score Heatmap for Sleep Quality and Physical Activity")
plt.xlabel("Physical Activity Level")
plt.ylabel("Sleep Quality Level")
plt.savefig('.//html_plots/anxiety_heatmap.png')
plt.show()
```

### Horizontal Bar Plot (Interactive): 
This is an interactive plot representing a horizontally stacked bar plot. This graph compares the following features in relation with anxiety score:
 - Social Support
 - Relationship Status
 - Substance Use
 - Counseling Service Use
 - Extracurricular Involvement
 
Users can view the different anxiety score distribution for these features by using the drop-down menu. 

```python
## Select the features that will be used to create the bar plot.
features = ["Social_Support", "Relationship_Status", "Substance_Use", "Counseling_Service_Use", "Extracurricular_Involvement"]

## Group the original DataFrame by the initalized features.
grouped_df = (
    df.melt(id_vars = ["Anxiety_Score"], value_vars = features, var_name = "Feature", value_name = "Level")
    .groupby(["Feature", "Level", "Anxiety_Score"])
    .size()
    .reset_index(name = "Count")
)

## Normalize the count for each anxiety score by dividing the raw count by total count of each category.
grouped_df["Normalized_Count"] = grouped_df.groupby(["Feature", "Level"])["Count"].transform(lambda x: x / x.sum())

## Intitialize the dropdown and selection menu on Altair.
dropdown = alt.binding_select(options = features, name = "Features: ")
selection = alt.selection_point(fields = ["Feature"], bind = dropdown, value = [{"Feature": "Social_Support"}])

## Initialize the horizontallys stacked bar plot with the dropdown menu.
base = alt.Chart(grouped_df).encode(
    y = alt.Y("Level:N", title = "Category"),
    x = alt.X("Count:Q", stack = "normalize", title = "Proportion by Anxiety Score"),
    color = alt.Color("Anxiety_Score:N", title = "Anxiety Score"),
    tooltip = [
        alt.Tooltip("Anxiety_Score:O", title = "Anxiety Score"),
        alt.Tooltip("Count:Q", title = "Count"),
        alt.Tooltip("Level:N", title = "Category"),
        alt.Tooltip("Feature:N", title = "Feature")
    ]
)

## Create the combined bar chart.
c1 = base.mark_bar().properties(
    height = 200 
)

## Use facet to simultaneously plot the bar plot for each category
bar_chart = c1.transform_filter(selection).facet(
    row = alt.Row("Feature:N", title = "Selected Feature")
).add_params(selection)

## Save and display the full bar plot.
bar_chart.save(".//html_plots//anxiety_bar_plot.html")
bar_chart
```
